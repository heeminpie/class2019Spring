```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
```

## 인터파크티켓에서 콘서트 항목의 예매랭킹 가져오기

```python
html = urlopen("http://ticket.interpark.com/contents/Ranking/RankList?pKind=01003&pCate=&pType=D&pDate=20190324")
bsObject = BeautifulSoup(html, "html.parser")
```

```python
bsObject
```

## n을 어떻게 가지고 와야할지 모르겠어요 

```python
rank = bsObject.find("div", "ranks").get_text()
print(int(rank)+1)
```

```python
for n in range(0,49):
    rank = bsObject.find("div", "ranks").get_text()
    div = bsObject.find("div",{"class":"prdInfo"}).find('b')
    title = div.get_text().strip()
    venue = div.next_sibling.strip()

    print(int(rank)+1, title, venue)
```

```python
with open('crawling.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['rank', 'title', 'venue'])
```
