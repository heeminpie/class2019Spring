```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation

import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = tensorflow.keras.datasets.boston_housing.load_data(path = 'boston_housing.npz', test_split=0.2, seed=113)
```

```python
model = Sequential()
model.add(Dense(1, input_shape=[13]))
#input13개 output1개
model.compile(optimizer='rmsprop', loss='mse') #Regression경우 rmsprop, mse/ classficiation은 또 다르다 후에 배울 것 
model.summary()
#결정해야할 거: hidden layers수, hidden layer의 node수, epochs 횟수(fit)
# activation(non-linearity부여는 hidden layer과 output layer에 부여가능, input layer에는 불가능)
#activation 종류:sigmoid(모든값을 0 or 1), tanh(모든 값을 -1 1 사이로 squash), =>남/녀 이런식으로 나누기 좋음
#relu(음값은 0, 플러스값은 그대로),softmax(node 총합이 1이되도록=>확률구하기 좋음)
#!!!!!!!!!!1 hidden layer에 주는 sigm, tanh, relu
#output layer에는 regression이면 아무것도x, classification이면 binary면 sigmoid, 여러개면 softmax

#시험: parameter개수 몇 개로 나오는가, 그림 어덯게 그려지는가 
```

```python
model.fit(x_train, y_train, epochs=1000)
#데이터 전체를 1번 돌림(epochs=1), 1000번 돌림 epochs=1000
```

```python
model.evaluate(x_test, y_test)
#fit=train
#evaluate=test
#predict=inference 실제 사용, 상품화
```

```python
#0514
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = tensorflow.keras.datasets.boston_housing.load_data(path = 'boston_housing.npz', test_split=0.2, seed=113)
```

```python
plt.plot(x_train[:,2], y_train,'.')
#x(0)는 crime, y는 집값
```

```python
boston_housing = ['crim', 'zn', 'indus', 'chas', 'nox','rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'black', 'lstat']
for i in range(13):
    plt.subplot(3, 5, i+1)
    plt.plot(x_train[:,i], y_train, '.')
    plt.title(boston_housing[i])
plt.show()
```

```python
model = Sequential()
model.add(Dense(10, input_shape=[13], activation = 'relu'))
model.add(Dense(1))
# input13개, hiddenlayernode10개, model.add(Dense(20))20개, 최종 출력 1개 => 히든레이어 총 두 개 
#시험: 레이어 몇 개 노드 몇 개 
# parm 140이 뭘까: parameter화살표 개수 동그라미 x1x2.....1 고로 13+1, 화살표는 14x10=140
#히든레이어도 1node필요, 근데 input에서 연결될 필요는 없다, 근데 그 다음 레이어(output)으로는 연결이 되어야해 그래서 (10+1)x1=11
model.compile(optimizer='rmsprop', loss='mse')
model.summary()
```

```python
model.fit(x_train, y_train, epochs=1000)
```

```python
#0521
print(model.input_shape)
print(model.output_shape)
#왜 (none, 13)으로 나오는가. 1x13, none=1, 하나의 샘플/ 근데 100x13으로 100개의 sample을 넣을 수도 있다, 고로 data개수로 변하니까 none으로
#정해지지 않았다(unknown)로 냅둔다, 다만 input, output에 해당하는 none은 개수가 같다
#(data의수가 많으면 계산되는 output도 그만큼 많아지니까)
#batch_size = 100x13에서 100에 해당 
```
